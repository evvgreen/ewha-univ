##클러스터링
: n개의 점을 k개의 그룹으로 나누고 각 그룹의 중심이 되는 k개의 점을 선택하는 문제.

단, 가장 큰 반경을 가진 그룹의 직경이 최소가 되도록 k개의 점을 선택

```python
import math, random

X = [[0, 1, 1], [1, 9, 8], [2, 2, 2], [3, 7, 2], [4, 8, 2],
     [5, 3, 4], [6, 1, 6], [7, 1, 9], [8, 7, 7], [9, 3, 8],
     [10, 6, 3], [11, 8, 6], [12, 5, 1], [13, 3, 2], [14, 3, 3],
     [15, 7, 3], [16, 4, 6], [17, 8, 8], [18, 1, 2], [10, 10, 10]]

K = 4
N = len(X)

isCenter = [False for _ in range(N)]
D = [[] for _ in range(N)]
C = []

i = random.randint(0, N-1)
C.append(X[i])
isCenter[i] = True
centerCnt = 1

while centerCnt < K:
    for i in range(N):
        if not isCenter[i]:
            D[i] = [-1, float('inf')]
            for j in range(centerCnt):
                dist = int(math.sqrt((C[j][1] - X[i][1])**2 + (C[j][2] - X[i][2])**2))
                if D[i][1] > dist:
                    D[i] = [j, dist]
    farthest = 0
    for i in range(N):
        if (not isCenter[i]) and (farthest < D[i][1]):
            farthest = D[i][1]
            nextCenter = i

    C.append(X[nextCenter])
    isCenter[nextCenter] = True
    centerCnt += 1
for i in range(K):
    print('Cluster %d : \t P. %d (%d, %d)' %(i, C[i][0], C[i][1], C[i][2]))
print('')

for i in range(N):
    if not isCenter[i]:
        D[i][1] = float('inf')
        for j in range(K):
            dist = math.sqrt((C[j][1]-X[i][1])**2 + (X[j][2]-X[i][2])**2)
            if D[i][1] > dist:
                D[i] = [j, dist]
        print('P. %d (%d, %d): \t Cluster %d, D. = %4.2f' %(X[i][0], X[i][1], X[i][2], D[i][0], D[i][1]))
```

##백트래킹

```python
result = []
prev_elements = []

def permute(elements):
    if len(elements) == 0:
        result.append(prev_elements[:])

    for e in elements:
        next_elements = elements[:]
        next_elements.remove(e)

        prev_elements.append(e)
        permute(next_elements)
        prev_elements.pop()

    return result

print(permute([0, 1, 2]))
```

##여왕 말 문제

```python
def is_threatening(i, j):
    for k in range(n):
        if board[i][k] == 1 or board[k][j] == 1:
            return True
    for k in range(n):
        for l in range(n):
            if (k+1 ==i+j) or (k-1 == i-j):
                if board[k][l] == 1:
                    return True

    return False

def queen(i):
    if i == n:
        return True
    for j in range(n):
        if not(is_threatening(i, j)) and (board[i][j] != 1):
            board[i][j] = 1
            if queen(i+1):
                return True
            board[i][j] = 0

    return False

n = 4
board = [[0 for _ in range(n)] for _ in range(n)]
queen(1)

for x in board:
    print(x, sep=' ')
```
